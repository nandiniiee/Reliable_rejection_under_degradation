```python
import pandas as pd

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
)
```

```python
# loading test predictions
df = pd.read_csv("test_predictions.csv")
print(df.shape)
df.head()
```

```python
# creating fixed threshold decisions
THRESHOLD = 0.60
df["threshold_decision"] = (df["best_similarity"] >= THRESHOLD).astype(int)
```

```python
# ground truth
y_true = df["label"]
y_pred_threshold = df["threshold_decision"]
```

```python
# threshold results
print("THRESHOLD METHOD")
print("Accuracy:", accuracy_score(y_true, y_pred_threshold))
print("Precision:", precision_score(y_true, y_pred_threshold))
print("Recall:", recall_score(y_true, y_pred_threshold))
print("F1:", f1_score(y_true, y_pred_threshold))
```

```python
# threshold confusion matrix
cm = confusion_matrix(y_true, y_pred_threshold)
print(cm)
print(classification_report(y_true, y_pred_threshold))
```

```python
# svm results
y_pred_svm = df["svm_decision"]
print("SVM METHOD")
print("Accuracy:", accuracy_score(y_true, y_pred_svm))
print("Precision:", precision_score(y_true, y_pred_svm))
print("Recall:", recall_score(y_true, y_pred_svm))
print("F1:", f1_score(y_true, y_pred_svm))
```

```python
# comparison table
comparison = pd.DataFrame(
    {
        "Method": ["Fixed Threshold", "SVM"],
        "Accuracy": [
            accuracy_score(y_true, y_pred_threshold),
            accuracy_score(y_true, y_pred_svm),
        ],
        "Precision": [
            precision_score(y_true, y_pred_threshold),
            precision_score(y_true, y_pred_svm),
        ],
        "Recall": [
            recall_score(y_true, y_pred_threshold),
            recall_score(y_true, y_pred_svm),
        ],
        "F1": [f1_score(y_true, y_pred_threshold), f1_score(y_true, y_pred_svm)],
    }
)

print(comparison)
```
