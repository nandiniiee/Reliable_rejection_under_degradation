```python
#creating clean data
import pandas as pd

clean_df = pd.read_csv("clean_testdata.csv")

clean_train = clean_df[
    [
        "quality_score",
        "det_score",
        "best_similarity",
        "margin",
        "label"
    ]
]

print(clean_train.shape)

clean_train.to_csv(
    "clean_train_features.csv",
    index=False
)
```

```python
#appending to the existing data
import pandas as pd

extended_df = pd.read_csv(
    "train_svm_extended.csv"
)

clean_df = pd.read_csv(
    "clean_train_features.csv"
)

new_training_df = pd.concat(
    [
        extended_df,
        clean_df
    ],
    ignore_index=True
)

print(new_training_df.shape)

print(
    new_training_df["label"]
    .value_counts()
)

new_training_df.to_csv(
    "train_svm_extended_all.csv",
    index=False
)
```

```python
#checking class (0 and 1) balance
print(
    new_training_df["label"]
    .value_counts()
)

print(
    new_training_df["label"]
    .value_counts(normalize=True)
    * 100
)
```
