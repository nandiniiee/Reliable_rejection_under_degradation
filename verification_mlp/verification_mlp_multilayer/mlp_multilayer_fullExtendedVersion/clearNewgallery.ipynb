```python
# Cell 1: Delete all image files inside gallery identity folders
import os

# Path to your TinyFace test directory
BASE_DIR = "/Users/admin/Desktop/reliable_rejection_under_degradation/verification_mlp/verification_mlp_multilayer/mlp_multilayer_fullExtendedVersion/tinyface_test"  # <-- Change this

GALLERY_DIR = os.path.join(BASE_DIR, "gallery")

# Image extensions to delete
IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp"}

deleted_count = 0

for identity in os.listdir(GALLERY_DIR):
    identity_path = os.path.join(GALLERY_DIR, identity)

    if not os.path.isdir(identity_path):
        continue

    for file in os.listdir(identity_path):
        file_path = os.path.join(identity_path, file)

        if (
            os.path.isfile(file_path)
            and os.path.splitext(file)[1].lower() in IMAGE_EXTENSIONS
        ):
            os.remove(file_path)
            deleted_count += 1

print(f"Done! Deleted {deleted_count} image(s).")
print("Identity folders have been preserved.")
```

```python
# verification if all gallery folders are empty now or not
# Check that all identity folders are now empty

import os

for identity in sorted(os.listdir(GALLERY_DIR)):
    identity_path = os.path.join(GALLERY_DIR, identity)

    if os.path.isdir(identity_path):
        remaining = os.listdir(identity_path)
        print(f"{identity}: {len(remaining)} file(s)")
```
