```python
# imports
import os
import random
import shutil
import pandas as pd

random.seed(42)

gallery_dir = "tinyface/Testing_Set/Gallery_Match"
probe_dir = "tinyface/Testing_Set/Probe"

OUTPUT_DIR = "tinyface_300"

gallery_out = os.path.join(OUTPUT_DIR, "gallery")

probe_out = os.path.join(OUTPUT_DIR, "probe")

os.makedirs(gallery_out, exist_ok=True)

os.makedirs(probe_out, exist_ok=True)

# -------------------------
# collect gallery images
# -------------------------

gallery_ids = {}

for file in os.listdir(gallery_dir):

    if not file.endswith(".jpg"):
        continue

    identity = file.split("_")[0]

    gallery_ids.setdefault(identity, []).append(file)

# -------------------------
# collect probe images
# -------------------------

probe_ids = {}

for file in os.listdir(probe_dir):

    if not file.endswith(".jpg"):
        continue

    identity = file.split("_")[0]

    probe_ids.setdefault(identity, []).append(file)

# -------------------------
# common identities
# -------------------------

common_ids = sorted(list(set(gallery_ids.keys()) & set(probe_ids.keys())))

print("Common identities:", len(common_ids))

# -------------------------
# select 300 identities
# -------------------------

selected_ids = random.sample(common_ids, 300)

print("Selected identities:", len(selected_ids))

# -------------------------
# copy files
# -------------------------

rows = []

for identity in selected_ids:

    gallery_file = random.choice(gallery_ids[identity])

    probe_file = random.choice(probe_ids[identity])

    shutil.copy(
        os.path.join(gallery_dir, gallery_file), os.path.join(gallery_out, gallery_file)
    )

    shutil.copy(
        os.path.join(probe_dir, probe_file), os.path.join(probe_out, probe_file)
    )

    rows.append(
        {
            "identity": identity,
            "gallery_image": gallery_file,
            "probe_image": probe_file,
        }
    )

pairs_df = pd.DataFrame(rows)

pairs_df.to_csv("tinyface_300_pairs.csv", index=False)

print()
print("Gallery images:", len(os.listdir(gallery_out)))

print("Probe images:", len(os.listdir(probe_out)))

print()
print(pairs_df.head())
```
