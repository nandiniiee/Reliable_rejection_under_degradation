```python
# imports
import os
import cv2
from pathlib import Path
```

```python
# paths
# Change this to your dataset path
BASE_DIR = "/Users/admin/Desktop/reliable_rejection_under_degradation/face_enhancement/initial_basics/nandini"

PROBE_DIR = os.path.join(BASE_DIR, "probe")
OUTPUT_DIR = os.path.join(BASE_DIR, "degraded_probes")

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# Gaussian Blur kernel for "High Blur"
BLUR_KERNEL = (31, 31)
SIGMA = 15

# Resize dimensions
RESIZE_DIM = (100, 100)

supported = (".jpg", ".jpeg", ".png")

for img_name in os.listdir(PROBE_DIR):

    if not img_name.lower().endswith(supported):
        continue

    img_path = os.path.join(PROBE_DIR, img_name)

    image = cv2.imread(img_path)

    if image is None:
        print(f"Could not read {img_name}")
        continue

    stem = Path(img_name).stem

    save_folder = os.path.join(OUTPUT_DIR, stem)
    os.makedirs(save_folder, exist_ok=True)

    ####################################################
    # High Blur
    ####################################################
    blur = cv2.GaussianBlur(image, BLUR_KERNEL, SIGMA)

    blur_path = os.path.join(save_folder, f"{stem}_blur_high.jpg")

    cv2.imwrite(blur_path, blur)

    ####################################################
    # Resize to 100x100
    ####################################################
    h, w = image.shape[:2]
    small = cv2.resize(image, (100, 100), interpolation=cv2.INTER_AREA)
    resize_degraded = cv2.resize(small, (w, h), interpolation=cv2.INTER_CUBIC)
    resize_path = os.path.join(save_folder, f"{stem}_resize100.jpg")
    cv2.imwrite(resize_path, resize_degraded)
    print(f"Process {img_name}")
```

```python
# verification
for root, dirs, files in os.walk(OUTPUT_DIR):
    level = root.replace(OUTPUT_DIR, "").count(os.sep)
    indent = " " * 4 * level
    print(f"{indent}{os.path.basename(root)}/")

    subindent = " " * 4 * (level + 1)
    for f in files:
        print(f"{subindent}{f}")
```
